```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df = pd.read_csv(r"C:\Users\AKHIL\OneDrive\Desktop\domestic_visitors.csv")
```

```python
df = df[['visitors', 'date', 'year','month','district']]
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
```

```python
df.head()
```

```python
df_yearly = df.groupby(['year', 'district']).sum()
```

```python
df_yearly = df_yearly.reset_index().pivot(index='year', columns='district', values='visitors')
```

```python
df_yearly.plot()
```

```python
df_yearly = df['visitors'].resample('Y').sum()
```

```python
df_yearly.plot()
```

```python
train = df_yearly[:'2019-12-31']
test = df_yearly['2020-01-01':]
```

```python
model = sm.tsa.arima.ARIMA(train, order=(1, 1, 1)).fit()
```

```python
pred = model.predict(start=test.index[0], end=test.index[-1])
```

```python
forecast = model.predict(start=pd.Timestamp('2025-01-01'), end=pd.Timestamp('2025-12-31'))
print("Projected number of domestic_visitors in Hyderabad in 2025:", int(forecast.sum()))
```
